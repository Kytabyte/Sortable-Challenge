```python
import json
import os
import sys
```

```python
products = []
with open('products.txt') as f:
    lines = f.readlines()
    for line in lines:
        products.append(json.loads(line))

listings = []
with open('listings.txt') as f:
    lines = f.readlines()
    for line in lines:
        listings.append(json.loads(line))
```

```python
len(products),len(listings)
```

```python
products[0], listings[0]
```

```python
type(products[0])
```

```python
manu = set()
for prod in products:
    manu.add(prod['manufacturer'].lower())
```

```python
manu
```

```python
extra_manu = set()
extra = 0
for listing in listings:
    if listing['manufacturer'] not in manu:
        extra += 1
        extra_manu.add(listing['manufacturer'])
```

```python
extra
```

```python
# check the first word
first_word_not_manuf_count = 0
first_word_not_manuf = dict()
for listing in listings:
    first_word = listing['title'].find(' ')
    if listing['title'][:first_word].lower() not in manu:
        first_word_not_manuf_count += 1
        if listing['title'][:first_word].lower() in first_word_not_manuf:
            first_word_not_manuf[listing['title'][:first_word].lower()] += 1
        else:
            first_word_not_manuf[listing['title'][:first_word].lower()] = 1
```

```python
first_word_not_manuf_count
```

```python
len(first_word_not_manuf)
```

```python
print(first_word_not_manuf)
```

```python
for x in sorted(list(map((lambda x : (first_word_not_manuf[x],x)), list(first_word_not_manuf)))):
    print(x)
```

```python
products_manu_model = {}
for prod in products:
    if prod['manufacturer'].lower() in products_manu_model:
        products_manu_model[prod['manufacturer'].lower()].append(prod['model'])
    else:
        products_manu_model[prod['manufacturer'].lower()] = [prod['model']]
```

```python
products_manu_model
```

```python
for entry in products_manu_model:
    print(entry, len(products_manu_model[entry]))
```
