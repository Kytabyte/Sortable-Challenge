```python
import json
import os
import sys
import re
```

```python
products = []
with open('products.txt') as f:
    lines = f.readlines()
    for line in lines:
        products.append(json.loads(line))

listings = []
with open('listings.txt') as f:
    lines = f.readlines()
    for line in lines:
        listings.append(json.loads(line))
```

```python
manu = set()
for prod in products:
    manu.add(prod['manufacturer'].lower())
```

```python
manu
```

```python
extra_manu = set()
extra = 0
for listing in listings:
    if listing['manufacturer'].lower() not in manu:
        extra += 1
        extra_manu.add(listing['manufacturer'].lower())
```

```python
extra, extra_manu
```

```python
# check the first word
first_word_not_manuf_count = 0
first_word_not_manuf = dict()
for listing in listings:
    first_word = listing['title'].find(' ')
    if listing['title'][:first_word].lower() not in manu:
        first_word_not_manuf_count += 1
        if listing['title'][:first_word].lower() in first_word_not_manuf:
            first_word_not_manuf[listing['title'][:first_word].lower()] += 1
        else:
            first_word_not_manuf[listing['title'][:first_word].lower()] = 1
```

```python
products_manu_model = {}
for prod in products:
    if prod['manufacturer'].lower() in products_manu_model:
        products_manu_model[prod['manufacturer'].lower()].append(prod['model'])
    else:
        products_manu_model[prod['manufacturer'].lower()] = [prod['model']]
```

```python
all_model = []
for entry in products_manu_model:
    all_model.extend(products_manu_model[entry])
print(sorted(all_model))
```

```python
# create a data structure to store result
result = {}
for prod in products:
    result[prod['manufacturer']] = [] # used to add listings
```

```python
# create a data structure to store products : {manufacturer:(family,model,name)}
products_dict = {}
for prod in products:
    if prod['manufacturer'].lower() not in products_dict:
        products_dict[prod['manufacturer'].lower()] = []
    products_dict[prod['manufacturer'].lower()].append((prod['family'] if 'family' in prod else '', prod['model'], prod['product_name']))
```

```python
# check the first word
first_word_not_manuf_count_re = 0
first_word_not_manuf_re = dict()
pattern = r'\b[^\d^\s^\W]+\b'
for listing in listings:
    try:
        first_word = re.search(pattern,listing['title']).group(0)
    except Exception as e:
        print(listing['title'])
    if first_word.lower() not in manu:
        first_word_not_manuf_count_re += 1
        if first_word.lower() in first_word_not_manuf_re:
            first_word_not_manuf_re[first_word.lower()] += 1
        else:
            first_word_not_manuf_re[first_word.lower()] = 1
```
